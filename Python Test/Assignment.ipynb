```python
import pandas as pd
Batting = pd.read_csv('Batting.csv')
print(Batting.head())
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()


ax.scatter(Batting['runs'], Batting['balls'])

ax.set_title('Batting Dataset')
ax.set_xlabel('Runs')
ax.set_ylabel('Balls')
```

```python
Batting.plot.scatter(x='runs', y='strike_rate', title='Best Batting Strike Rate')
```

```python

columns = Batting.columns.drop(['minutes','dismissal','player','team','opponent','innings','match_id'])

x_data = range(0, Batting.shape[0])

fig, ax = plt.subplots()

for column in columns:
    ax.plot(x_data, Batting[column])

ax.set_title('Batting Dataset')
ax.legend()
```

```python


Batting.plot.hist(subplots=True, layout=(8,8), figsize=(50, 50), bins=20)
```
